###**Visualization**

```python
sns.histplot(data = life_index, x = life_index['  Life satisfaction'], kde = True)
plt.title('Better Life Index 2024 - Life Satisfaction Distribution')
plt.show()
```

```python
plt.figure(figsize = (14, 8))
sns.heatmap(data = life_index.drop(columns = 'Country').corr(), cmap = 'coolwarm', annot = True, fmt = '.1f')
plt.title('C͟o͟r͟r͟e͟l͟a͟t͟i͟o͟n͟ ͟H͟e͟a͟t͟m͟a͟p͟ ͟V͟i͟s͟u͟a͟l͟i͟z͟e͟d͟', fontsize = 20, pad = 20)
plt.xlabel('Features', fontsize = 15)
plt.ylabel('Features', fontsize = 15)
plt.show()
```
